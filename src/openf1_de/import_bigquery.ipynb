```python
from google.cloud import bigquery

client = bigquery.Client(project="trial-f1", location="asia-southeast2")

uri = "gs://data-mart-f1/driver_standings/2026_verify.json"  # single file

job = client.extract_table(
    "trial-f1.openf1_marts.mart_drivers",
    uri,
    job_config=bigquery.ExtractJobConfig(
        destination_format=bigquery.DestinationFormat.NEWLINE_DELIMITED_JSON,
    ),
)
job.result()
print("Wrote", uri)
```
